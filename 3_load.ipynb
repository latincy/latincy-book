## Loading LatinCy models with spaCy

```python
# Imports

import spacy
from pprint import pprint
```

```python
nlp = spacy.load('la_core_web_lg')
```

## Creating a spaCy doc from a string

```python
text = "Haec narrantur a poetis de Perseo."
doc = nlp(text)
print(doc)
```

```python
print(type(doc))
```

```python
print(doc.__repr__())
```

```python
print(doc.text)
```
